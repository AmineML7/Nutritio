Recuperation Data micronutriments sur aliments

```python
aliments = pd.read_excel("Data/Aliments/Recherche/TABLE_CIQUAL_2020.xls")
aliments.head()
```

```python
colonnes_cles = [
    # Identifiants et catégories
    'alim_code', 
    'alim_nom_fr',
    'alim_grp_nom_fr', 
    'alim_ssgrp_nom_fr',
    
    # Macronutriments essentiels (une seule version)
    'Energie, Règlement UE N° 1169/2011 (kcal/100 g)',
    'Protéines, N x facteur de Jones (g/100 g)',
    'Glucides (g/100 g)',
    'Lipides (g/100 g)',
    'Fibres alimentaires (g/100 g)',
    
    # Minéraux importants
    'Calcium (mg/100 g)',
    'Fer (mg/100 g)',
    'Magnésium (mg/100 g)',
    'Potassium (mg/100 g)',
    'Zinc (mg/100 g)',
    
    # Vitamines
    'Rétinol (µg/100 g)',
    'Beta-Carotène (µg/100 g)',
    'Vitamine D (µg/100 g)',
    'Vitamine E (mg/100 g)',
    'Vitamine C (mg/100 g)',
    'Vitamine B1 ou Thiamine (mg/100 g)',
    'Vitamine B2 ou Riboflavine (mg/100 g)',
    'Vitamine B9 ou Folates totaux (µg/100 g)',
    'Vitamine B12 (µg/100 g)'
]
```

```python
aliments_structures = aliments[colonnes_cles]
aliments_structures.head()
```

```python
aliments_structures = aliments_structures.drop(0).reset_index(drop=True)
```

```python
aliments_structures.columns
```

```python
new_column_names = {
    'alim_code': 'code_aliment',
    'alim_nom_fr': 'nom_aliment',
    'alim_grp_nom_fr': 'groupe_alimentaire',
    'alim_ssgrp_nom_fr': 'sous_groupe_alimentaire',
    'Energie, Règlement UE N° 1169/2011 (kcal/100 g)': 'energie_kcal',
    'Protéines, N x facteur de Jones (g/100 g)': 'proteines',
    'Glucides (g/100 g)': 'glucides',
    'Lipides (g/100 g)': 'lipides',
    'Fibres alimentaires (g/100 g)': 'fibres',
    'Calcium (mg/100 g)': 'calcium',
    'Fer (mg/100 g)': 'fer',
    'Magnésium (mg/100 g)': 'magnesium',
    'Potassium (mg/100 g)': 'potassium',
    'Zinc (mg/100 g)': 'zinc',
    'Rétinol (µg/100 g)': 'retinol',
    'Beta-Carotène (µg/100 g)': 'beta_carotene',
    'Vitamine D (µg/100 g)': 'vitamine_D',
    'Vitamine E (mg/100 g)': 'vitamine_E',
    'Vitamine C (mg/100 g)': 'vitamine_C',
    'Vitamine B1 ou Thiamine (mg/100 g)': 'vitamine_B1',
    'Vitamine B2 ou Riboflavine (mg/100 g)': 'vitamine_B2',
    'Vitamine B9 ou Folates totaux (µg/100 g)': 'vitamine_B9',
    'Vitamine B12 (µg/100 g)': 'vitamine_B12'
}
```

```python
aliments_structures = aliments_structures.rename(columns=new_column_names)
```

```python
import numpy as np

nutrient_cols = [
    'energie_kcal', 'proteines', 'glucides', 'lipides', 'fibres',
    'calcium', 'fer', 'magnesium', 'potassium', 'zinc',
    'retinol', 'beta_carotene', 'vitamine_D', 'vitamine_E',
    'vitamine_C', 'vitamine_B1', 'vitamine_B2', 'vitamine_B9', 'vitamine_B12'
]
```

```python
import pandas as pd

def clean_nutrient_value(x):
    """Convertit les valeurs textuelles en nombres"""
    if pd.isna(x) or x == '-':
        return np.nan
    if isinstance(x, str):
        # Gestion des valeurs avec '<' (prend la moitié de la limite)
        if x.startswith('<'):
            x = x.replace('<', '').strip()
        # Gestion des valeurs avec 'j' (jour) - exemple: "0 j"
        if 'j' in x:
            x = x.split()[0]
        # Remplacement des virgules par points pour les nombres
        x = x.replace(',', '.')
    try:
        return float(x)
    except (ValueError, TypeError):
        return np.nan


for col in nutrient_cols:
    aliments_structures[col] = aliments_structures[col].apply(clean_nutrient_value)

# Vérification
aliments_structures[nutrient_cols].head()
```

```python
aliments_structures.describe()
```

```python
aliments_structures['complet'] = aliments_structures[nutrient_cols].notna().all(axis=1)

resultat = aliments_structures.groupby('groupe_alimentaire')['complet'].agg(
    total_aliments='count',
    aliments_complets='sum',
    pourcentage_complets=lambda x: round(x.mean()*100, 1)
).sort_values('aliments_complets', ascending=False)

print("Aliments complets par groupe :")
resultat
```
